```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/war-points-folder/chap_6_slayed_points.csv')

point = df['score']
x_axis = range(len(point))
y_axis = point
```

```python
import matplotlib.pyplot as plt

plt.plot(x_axis,y_axis)
plt.title('Bhishma Parva')
plt.xlabel('Time')
plt.ylabel('Pandavas Performance')
plt.show()
```
